# Midnight Snack Choices and Productivity Analysis
This notebook analyzes the relationship between midnight snack choices, productivity, and meme-sharing abilities of students. We will use the provided datasets to explore correlations between snack types, productivity levels, and energy.

### Guiding Questions:
- How do different midnight snack choices correlate with next-day productivity?
- Is there a link between snack quantity and meme-sharing creativity?
- Which snack types have the highest impact on energy levels?
- What snack recommendations should be given to students for balancing productivity and meme quality?


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
students_df = pd.read_csv('data/students_snack.csv')
snack_consumption_df = pd.read_csv('data/snack_consumption.csv')
productivity_df = pd.read_csv('data/productivity.csv')

# Display first few rows of each dataset
display(students_df.head())
display(snack_consumption_df.head())
display(productivity_df.head())
```

## 1. How do different midnight snack choices correlate with next-day productivity?
Here, we explore how snack types like pizza, chips, fruit, etc., affect next-day study hours and energy levels.


```python
# Merge snack consumption and productivity data
merged_data = pd.merge(snack_consumption_df, productivity_df, on=['student_id', 'date'])
merged_data = pd.merge(merged_data, students_df[['student_id', 'favorite_snack']], on='student_id')

# Plot hours studied vs favorite snack
plt.figure(figsize=(10, 6))
sns.boxplot(x='favorite_snack', y='hours_studied', data=merged_data)
plt.title('Hours Studied by Favorite Snack')
plt.xlabel('Favorite Snack')
plt.ylabel('Hours Studied')
plt.show()
```

## 2. Is there a link between snack quantity and meme-sharing creativity?
Here, we investigate whether students who consume more snacks are more creative in meme-sharing.


```python
# Scatter plot of meme-sharing vs snack quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='quantity', y='memes_shared', data=merged_data, hue='snack_type')
plt.title('Meme-Sharing vs Snack Quantity')
plt.xlabel('Snack Quantity')
plt.ylabel('Memes Shared')
plt.legend(title='Snack Type')
plt.show()
```

## 3. Which snack types have the highest impact on energy levels?
In this section, we analyze the effect of different snack types on next-day energy levels.


```python
# Plot energy levels vs favorite snack
plt.figure(figsize=(10, 6))
sns.boxplot(x='favorite_snack', y='energy_level', data=merged_data)
plt.title('Energy Level by Favorite Snack')
plt.xlabel('Favorite Snack')
plt.ylabel('Energy Level')
plt.show()
```

## 4. Recommendations for Snack Choices
Based on the analysis, we can make snack recommendations to students to balance productivity and meme quality.


- **Fruits and Nuts**: Best for maintaining high energy and good productivity, with moderate meme output.
- **Pizza and Chips**: Best for boosting meme creativity, but energy and productivity may take a hit.
- **Chocolate**: Offers a good middle ground between energy, productivity, and meme creation.
